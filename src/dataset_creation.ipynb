```python
import numpy as np

from generate_data import create_dataset
```

```python
# --- Dataset Generation ---
SCENARIO_NAMES = ["city_6_miami"]
BS_IDXS = [1, 2, 3]
N_SAMPLES = 10000
N_USERS = 16

PATCH_ROWS = 16
PATCH_COLS = 1

P_TOTAL = 1.0
SIGMA2 = 1e-3

SAVE = True

if SAVE:
    import os
    from faker import Faker
    
    DATA_DIR = "data/sim_" + "_".join(SCENARIO_NAMES) + f"_{Faker("de_DE").city().lower()}"

all_tokens_tensor, all_channels_array = create_dataset(scenario_names=SCENARIO_NAMES,
                                                       bs_idxs=BS_IDXS,
                                                       n_samples=N_SAMPLES,
                                                       n_users=N_USERS,
                                                       patch_rows=PATCH_ROWS,
                                                       patch_cols=PATCH_COLS)

_, N, _, M = all_channels_array.shape
```

```python
if SAVE:

    if not os.path.exists(f"../{DATA_DIR}"):
        os.makedirs(f"../{DATA_DIR}")
        
    np.save(f"../{DATA_DIR}/channels.npy", all_channels_array)

    parameters = {
        "scenario_names": SCENARIO_NAMES,
        "bs_idxs": BS_IDXS,
        "samples": N_SAMPLES,
        "users": N_USERS,
        "subcarriers": N,
        "bs_antennas": M,
        "p_total": P_TOTAL,
        "sigma2": SIGMA2,
        # "patch_rows": PATCH_ROWS,
        # "patch_cols": PATCH_COLS
    }

    with open(f"../{DATA_DIR}/parameters.txt", "w") as f:
        for key, value in parameters.items():
            f.writelines(f"{key}: {value}\n")
```
